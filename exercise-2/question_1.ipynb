```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
u = np.arange(0, 1.01, 0.01)
p0 = np.array([4, 4])
p1 = np.array([6, 8])
p2 = np.array([13, 5])
```

```python
p = lambda u: (1 - u) ** 2 * p0 + .5 * u * (4 - 3 * u) * p1 + .5 * u ** 2 * p2
b = lambda u: (1 - u) ** 2 * p0 + 2 * u * (1 - u) * p1 + u ** 2 * p2
```

```python
x1, y1 = [], []
for x_t, y_t in map(p, u):
    x1.append(x_t)
    y1.append(y_t)
    
x2, y2 = [], []
for x_t, y_t in map(b, u):
    x2.append(x_t)
    y2.append(y_t)
```

```python
line1, = plt.plot(x1, y1, label="original")
line2, = plt.plot(x2, y2, label="bezier")

plt.legend(handles=[line1, line2])
```

```python
for x_t, y_t in zip(x2, y2):
    print(x_t, y_t)
```
